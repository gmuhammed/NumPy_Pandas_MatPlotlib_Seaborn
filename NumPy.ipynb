```python
xz = np.arange(0,20).reshape(4,5)
xz
```

```python
%%timeit
result=[]
for a in xz:
    for i in a:
        result.append(i)
result
#this is to show that we will need multiple for loop to iterate through multidimentional array
```

```python
%%timeit
result=[]
for i in xz.flatten():
    result.append(i)
result
```

```python
%%timeit
#nditer allow us easily access each element of multidimentional array 
#it is highly optimized for this sort of iteratio
result=[]
for i in np.nditer(xz):
    result.append(i)
result
```

```python
#ndenumerate # simlar to python enumerate that returns index,element pair
result=[]
for i,j in np.ndenumerate(xz):
    result.append([i,j])
result[:3]
```
